#Handling Categorical and Payment Status Variables

```python
import pandas as pd
import numpy as np




# Create a copy to avoid SettingWithCopyWarning during transformations
df_train_processed = df_train.copy()
df_validation_processed = df_validation.copy()

# 1. Refine Categorical Variables (One-Hot Encoding)
print("\n--- Refining Categorical Variables (One-Hot Encoding) ---")

# Define categorical features that should be one-hot encoded
# We use the _mapped columns for education, sex, marriage for better interpretability
# and include a mapping for 'Other' for sex, although it only has 0 and 1 in the data
categorical_cols_to_encode_train = ['marriage_mapped', 'sex_mapped', 'education_mapped']

# For validation set, ensure consistency in column names for mapping
# The mapping applied to df_train was not applied inplace to df_validation in previous step,
# so we re-apply it here for consistency before OHE.
df_validation_processed['marriage_mapped'] = df_validation_processed['marriage'].map({1: 'Married', 2: 'Single', 3: 'Others'}).fillna('Unknown_Code')
df_validation_processed['sex_mapped'] = df_validation_processed['sex'].map({0: 'Female', 1: 'Male'}).fillna('Unknown_Code')
df_validation_processed['education_mapped'] = df_validation_processed['education'].map({1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others', 5: 'Unknown', 6: 'Unknown', 0: 'Unknown'}).fillna('Unknown_Code')

# Perform One-Hot Encoding
# Use pd.get_dummies, ensuring consistent columns across train and validation sets
# by passing all unique categories found across both if needed, or by unioning columns.
# For simplicity, for now, we'll let get_dummies handle it per df and then align.
# Later, a robust approach might use a single encoder fitted on train and transformed on both.

df_train_encoded = pd.get_dummies(df_train_processed, columns=categorical_cols_to_encode_train, prefix=categorical_cols_to_encode_train, drop_first=False)
df_validation_encoded = pd.get_dummies(df_validation_processed, columns=categorical_cols_to_encode_train, prefix=categorical_cols_to_encode_train, drop_first=False)

print(f"Shape of df_train after OHE: {df_train_encoded.shape}")
print(f"Shape of df_validation after OHE: {df_validation_encoded.shape}")


# 2. Transform Payment Status Variables (pay_0 to pay_6)
print("\n--- Transforming Payment Status Variables (pay_m) ---")

pay_status_features = ['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']

# Create new features based on payment status categories
def map_pay_status(status):
    if status == -2:
        return 'No_Consumption'
    elif status == -1:
        return 'Paid_On_Time'
    elif status == 0:
        return 'Min_Payment_Made'
    elif status >= 1:
        return 'Overdue'
    else: # Handle any other unexpected values if they exist, though not seen in previous output
        return 'Unknown_Status'

for df in [df_train_encoded, df_validation_encoded]:
    for col in pay_status_features:
        df[f'{col}_status'] = df[col].apply(map_pay_status)

# One-hot encode the new _status columns
pay_status_cols_to_encode = [f'{col}_status' for col in pay_status_features]
df_train_encoded = pd.get_dummies(df_train_encoded, columns=pay_status_cols_to_encode, prefix=pay_status_cols_to_encode, drop_first=False)
df_validation_encoded = pd.get_dummies(df_validation_encoded, columns=pay_status_cols_to_encode, prefix=pay_status_cols_to_encode, drop_first=False)

print(f"Shape of df_train after pay_m status OHE: {df_train_encoded.shape}")
print(f"Shape of df_validation after pay_m status OHE: {df_validation_encoded.shape}")


# Create summary features from original pay_m values (before OHE of original columns)
for df in [df_train_encoded, df_validation_encoded]:
    # Number of delinquencies (months where pay_m >= 1)
    df['num_delinquencies'] = (df[pay_status_features] >= 1).sum(axis=1)

    # Max consecutive delinquencies (simple approach for now, can be refined)
    # A simple approach for max consecutive delinquencies
    # This involves iterating or using a rolling sum.
    # For now, let's keep it simple: any month with pay_m >= 1 implies a delinquency.
    # A more advanced feature engineering would be needed for a precise 'streak'.
    # For demonstration, let's just count consecutive 'overdue' statuses.
    # Create binary columns for 'overdue' status
    for col in pay_status_features:
        df[f'{col}_overdue'] = (df[col] >= 1).astype(int)

    # Sum of overdue months as a simple proxy for 'streak'
    df['total_overdue_months'] = df[[f'{col}_overdue' for col in pay_status_features]].sum(axis=1)

    # For consecutive overdue months, a loop is more robust
    def calculate_max_consecutive_overdue(row, pay_cols):
        max_streak = 0
        current_streak = 0
        for col in pay_cols:
            if row[col] >= 1:
                current_streak += 1
            else:
                max_streak = max(max_streak, current_streak)
                current_streak = 0
        return max(max_streak, current_streak) # Return max of current_streak or final max_streak

    df['max_consecutive_overdue'] = df.apply(lambda row: calculate_max_consecutive_overdue(row, pay_status_features), axis=1)

    # Drop the temporary _overdue columns
    df.drop(columns=[f'{col}_overdue' for col in pay_status_features], inplace=True)


print("\nNew features derived from pay_m:")
print("df_train_encoded columns:", [col for col in df_train_encoded.columns if 'num_delinquencies' in col or 'total_overdue_months' in col or 'max_consecutive_overdue' in col])
print("df_validation_encoded columns:", [col for col in df_validation_encoded.columns if 'num_delinquencies' in col or 'total_overdue_months' in col or 'max_consecutive_overdue' in col])

# Display a sample of the processed dataframes to see new columns
print("\nFirst 5 rows of df_train_encoded with new categorical and pay_m features:")
print(df_train_encoded.head())

```

#Advanced Feature Engineering, Column Alignment, and Cleanup

```python
import pandas as pd
import numpy as np

# Assuming df_train_encoded and df_validation_encoded are available from Step 3.1


# Ensure we're working on copies to avoid modifying original df_train/df_validation
df_train_final = df_train_encoded.copy()
df_validation_final = df_validation_encoded.copy()

# Store Customer_ID for final submission and drop from features
train_customer_ids = df_train_final['Customer_ID']
validation_customer_ids = df_validation_final['Customer_ID']

# Drop Customer_ID from both dataframes as it's not a feature
df_train_final.drop('Customer_ID', axis=1, inplace=True)
df_validation_final.drop('Customer_ID', axis=1, inplace=True)

# Define columns that are 'original' and now redundant for feature engineering
original_categorical_cols = ['marriage', 'sex', 'education', 'marriage_mapped', 'sex_mapped', 'education_mapped']
pay_m_cols = ['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
bill_amt_cols = ['Bill_amt1', 'Bill_amt2', 'Bill_amt3', 'Bill_amt4', 'Bill_amt5', 'Bill_amt6']
pay_amt_cols = ['pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']

# Combine all columns to drop
columns_to_drop = original_categorical_cols + pay_m_cols + bill_amt_cols + pay_amt_cols

# Filter out columns that might not exist in case they were already dropped or renamed
columns_to_drop_train = [col for col in columns_to_drop if col in df_train_final.columns]
columns_to_drop_validation = [col for col in columns_to_drop if col in df_validation_final.columns]


# 1. Advanced Feature Engineering (Credit Utilization and Bill Trends)
print("\n--- Advanced Feature Engineering ---")

# Credit Utilization Ratios
for i in range(1, 7):
    # Ensure no division by zero for LIMIT_BAL
    df_train_final[f'utilization_m{i}'] = np.where(df_train_final['LIMIT_BAL'] == 0, 0, df_train_final[f'Bill_amt{i}'] / df_train_final['LIMIT_BAL'])
    df_validation_final[f'utilization_m{i}'] = np.where(df_validation_final['LIMIT_BAL'] == 0, 0, df_validation_final[f'Bill_amt{i}'] / df_validation_final['LIMIT_BAL'])

# Aggregate Credit Utilization (e.g., average, max over 6 months)
utilization_cols = [f'utilization_m{i}' for i in range(1, 7)]
df_train_final['avg_utilization'] = df_train_final[utilization_cols].mean(axis=1)
df_train_final['max_utilization'] = df_train_final[utilization_cols].max(axis=1)
df_validation_final['avg_utilization'] = df_validation_final[utilization_cols].mean(axis=1)
df_validation_final['max_utilization'] = df_validation_final[utilization_cols].max(axis=1)

# Bill Amount Changes (month-over-month)
# Bill_amt1 (most recent) to Bill_amt6 (oldest)
for i in range(1, 6): # Loop from 1 to 5 to get 5 differences
    df_train_final[f'Bill_change_m{i}-{i+1}'] = df_train_final[f'Bill_amt{i}'] - df_train_final[f'Bill_amt{i+1}']
    df_validation_final[f'Bill_change_m{i}-{i+1}'] = df_validation_final[f'Bill_amt{i}'] - df_validation_final[f'Bill_amt{i+1}']

# Payment Amount Changes (month-over-month)
for i in range(1, 6): # Loop from 1 to 5 to get 5 differences
    df_train_final[f'Pay_change_m{i}-{i+1}'] = df_train_final[f'pay_amt{i}'] - df_train_final[f'pay_amt{i+1}']
    df_validation_final[f'Pay_change_m{i}-{i+1}'] = df_validation_final[f'pay_amt{i}'] - df_validation_final[f'pay_amt{i+1}']


# 2. Handle Extreme Values in PAY_TO_BILL_ratio
print("\n--- Handling Extreme Values in PAY_TO_BILL_ratio ---")
# Using 1st and 99th percentile for winsorization
# Calculate percentiles only from training data to avoid data leakage
lower_bound_train = df_train_final['PAY_TO_BILL_ratio'].quantile(0.01)
upper_bound_train = df_train_final['PAY_TO_BILL_ratio'].quantile(0.99)

df_train_final['PAY_TO_BILL_ratio'] = np.clip(df_train_final['PAY_TO_BILL_ratio'], lower_bound_train, upper_bound_train)
df_validation_final['PAY_TO_BILL_ratio'] = np.clip(df_validation_final['PAY_TO_BILL_ratio'], lower_bound_train, upper_bound_train)

print(f"PAY_TO_BILL_ratio clipped to range [{lower_bound_train:.2f}, {upper_bound_train:.2f}] based on training data.")


# 3. Drop Redundant Columns
print("\n--- Dropping Redundant Columns ---")
df_train_final.drop(columns=columns_to_drop_train, inplace=True, errors='ignore')
df_validation_final.drop(columns=columns_to_drop_validation, inplace=True, errors='ignore')


# 4. Column Alignment for Train and Validation Sets
# This is critical to ensure both dataframes have the exact same columns before modeling.
print("\n--- Aligning Columns of Training and Validation Sets ---")
train_cols = set(df_train_final.columns)
# Exclude the target variable from validation columns if it exists
validation_cols = set(df_validation_final.columns)

# Make sure 'next_month_default' is not in validation_cols for unioning features
if 'next_month_default' in validation_cols:
    validation_cols.remove('next_month_default')

# Union of all feature columns (excluding target)
all_feature_cols = sorted(list(train_cols.union(validation_cols) - {'next_month_default'}))

# Reindex and fill missing columns with 0.
# For training set, add back 'next_month_default' if it was removed in the unioning step
# This ensures target column is present and is at the end.
target_col_train = df_train_final['next_month_default']
df_train_final = df_train_final[all_feature_cols]
df_train_final['next_month_default'] = target_col_train

df_validation_final = df_validation_final.reindex(columns=all_feature_cols, fill_value=0)

print(f"Final shape of df_train_final: {df_train_final.shape}")
print(f"Final shape of df_validation_final: {df_validation_final.shape}")
print(f"Number of columns match: {df_train_final.drop('next_month_default', axis=1).shape[1] == df_validation_final.shape[1]}")

print("\nFirst 5 rows of df_train_final (after all preprocessing):")
print(df_train_final.head())
print("\nFirst 5 rows of df_validation_final (after all preprocessing):")
print(df_validation_final.head())


print("\n--- Step 3.2 Complete ---")

# Store the processed dataframes globally for next steps
# global df_train_final, df_validation_final, train_customer_ids, validation_customer_ids
# df_train_final = df_train_final
# df_validation_final = df_validation_final
# train_customer_ids = train_customer_ids
# validation_customer_ids = validation_customer_ids
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, make_scorer, fbeta_score
from sklearn.model_selection import cross_val_score, StratifiedKFold # For robustness

# Ensure X_train_resampled, y_train_resampled, X_test, y_test are available from the previous step
# For local testing, ensure these variables are in your environment from the previous successful run.


# Define models to train
models = {
    'Logistic Regression': LogisticRegression(random_state=42, solver='liblinear', max_iter=1000), # liblinear is good for small datasets, allows L1/L2
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'), # Suppress XGBoost deprecation warning
    'LightGBM': LGBMClassifier(random_state=42, verbose=-1) # verbose=-1 suppresses output
}

# Define evaluation metrics. F2-score is important for this project.
# We need to define a custom scorer for F2-score for cross_val_score.
# F-beta score: beta > 1 gives more weight to recall (identifying positives, i.e., defaulters)
# beta = 2 means recall is twice as important as precision.
f2_scorer = make_scorer(fbeta_score, beta=2)

results = {}

# Train and evaluate each model
for name, model in models.items():
    print(f"\n--- Training {name} ---")
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] # Probability of the positive class (1)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    f2 = fbeta_score(y_test, y_pred, beta=2) # Calculate F2-score
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'F2-Score': f2,
        'AUC-ROC': roc_auc
    }
    print(f"{name} Metrics on Test Set (after training on SMOTE data):")
    for metric, value in results[name].items():
        print(f"  {metric}: {value:.4f}")

    # Optionally, perform cross-validation on the original imbalanced training data for more robust estimation
    # This gives a baseline performance on realistic distribution for comparison.
    # Note: Cross-validation on *resampled* data can be misleading due to data leakage if not done carefully.
    # For a fair baseline on the original distribution, we use the original X, y before SMOTE.
    # We will use the non-resampled X_train, y_train for cross-validation to simulate original imbalance for comparison.
    print(f"\n--- Cross-validation for {name} on (original, non-resampled) training data ---")
    # Using the X_train, y_train directly before SMOTE (the 80% split of original data)
    cv_scores = cross_val_score(model, X_train, y_train, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state), scoring=f2_scorer)
    print(f"  Cross-validation F2-Score (on non-resampled X_train): {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")

print("\n--- Summary of All Models' Initial Performance ---")
results_df = pd.DataFrame(results).T
print(results_df.round(4))

```

#XGBoost Tuning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, make_scorer, fbeta_score, precision_recall_curve, confusion_matrix, classification_report

# Assume X_train_resampled_scaled, y_train_resampled, X_test_scaled, y_test
# and X_train_scaled_for_cv, y_train are available from previous steps.

# Re-define f2_scorer
f2_scorer = make_scorer(fbeta_score, beta=2)
random_state = 42 # For reproducibility


# Calculate scale_pos_weight based on original y_train (non-resampled)
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
initial_scale_pos_weight = neg_count / pos_count
print(f"\nCalculated initial scale_pos_weight: {initial_scale_pos_weight:.2f} (neg_count={neg_count}, pos_count={pos_count})")

# Define models to tune and their parameter grids (ONLY XGBoost)
models_and_grids = {
    'XGBoost': {
        'model': XGBClassifier(random_state=random_state, use_label_encoder=False, eval_metric='logloss'),
        'params': {
            'n_estimators': [150, 300, 500],
            'learning_rate': [0.05, 0.1],
            'max_depth': [5, 7],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0],
            'scale_pos_weight': [1, initial_scale_pos_weight, initial_scale_pos_weight*1.5] # Include original and slightly higher weights
        }
    }
}

best_models_tuned = {}
tuned_metrics_default_threshold = {}

# 1. Hyperparameter Tuning (ONLY XGBoost)
for name, config in models_and_grids.items(): # This loop will now run only once for XGBoost
    print(f"\n--- Tuning {name} ---")
    model = config['model']
    param_grid = config['params']

    grid_search = GridSearchCV(estimator=model,
                               param_grid=param_grid,
                               scoring=f2_scorer, # Optimize for F2-Score
                               cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=random_state), # CV=3 for faster execution
                               verbose=1,
                               n_jobs=-1)

    # Fit on the SMOTE-resampled and scaled training data
    grid_search.fit(X_train_resampled_scaled, y_train_resampled)
    best_model_tuned = grid_search.best_estimator_
    best_score_cv = grid_search.best_score_

    best_models_tuned[name] = best_model_tuned
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best F2-Score on resampled training data (CV): {best_score_cv:.4f}")

    # Evaluate tuned model on X_test_scaled with default 0.5 threshold
    y_pred_proba_tuned = best_model_tuned.predict_proba(X_test_scaled)[:, 1]
    y_pred_tuned_default = (y_pred_proba_tuned >= 0.5).astype(int)

    tuned_metrics_default_threshold[name] = {
        'Accuracy': accuracy_score(y_test, y_pred_tuned_default),
        'Precision': precision_score(y_test, y_pred_tuned_default, zero_division=0),
        'Recall': recall_score(y_test, y_pred_tuned_default, zero_division=0),
        'F1-Score': f1_score(y_test, y_pred_tuned_default, zero_division=0),
        'F2-Score (Default Threshold)': fbeta_score(y_test, y_pred_tuned_default, beta=2, zero_division=0),
        'AUC-ROC': roc_auc_score(y_test, y_pred_proba_tuned)
    }
    print(f"{name} Metrics on Test Set (Tuned, Default Threshold 0.5):")
    for metric, value in tuned_metrics_default_threshold[name].items():
        print(f"  {metric}: {value:.4f}")

print("\n--- Summary of All Tuned Models' Performance (Default Threshold) ---")
# This table will now only contain XGBoost
tuned_results_df = pd.DataFrame(tuned_metrics_default_threshold).T
print(tuned_results_df.round(4))

# 2. Classification Threshold Tuning for XGBoost
print("\n--- Classification Threshold Tuning ---")

# Since we only tuned XGBoost, it is automatically the final_best_model_for_prediction
final_best_model_for_prediction = best_models_tuned['XGBoost'] # Directly select XGBoost
best_model_name_for_threshold = 'XGBoost' # Set name explicitly

print(f"\nProceeding with threshold tuning for: {best_model_name_for_threshold}")

# Get probabilities from the selected best model on the test set
y_pred_proba_final_model_test = final_best_model_for_prediction.predict_proba(X_test_scaled)[:, 1]

# --- DIAGNOSTIC: Analyze probability distribution ---
print("\n--- Diagnostic: Test Probability Distribution ---")
print(pd.Series(y_pred_proba_final_model_test).describe())
plt.figure(figsize=(8, 4))
sns.histplot(y_pred_proba_final_model_test, bins=50, kde=True)
plt.title('Distribution of Predicted Probabilities on Test Data')
plt.xlabel('Probability')
plt.ylabel('Count')
plt.show()
print("---------------------------------------")


# Explore thresholds to find the optimal F2-Score
thresholds = np.arange(0.01, 1.0, 0.005) # Test thresholds from 0.01 to 0.995 with finer steps
f2_scores_at_thresholds = []
for t in thresholds:
    y_pred_at_t = (y_pred_proba_final_model_test >= t).astype(int)
    if np.sum(y_pred_at_t) == 0 and np.sum(y_test) > 0: # If no positives predicted but actual positives exist
        f2_scores_at_thresholds.append(0.0)
    else:
        f2_scores_at_thresholds.append(fbeta_score(y_test, y_pred_at_t, beta=2, zero_division=0))

optimal_threshold_idx = np.argmax(f2_scores_at_thresholds)
optimal_threshold = thresholds[optimal_threshold_idx]
max_f2_score = f2_scores_at_thresholds[optimal_threshold_idx]

print(f"Optimal threshold for {best_model_name_for_threshold} (maximizing F2-Score on test set): {optimal_threshold:.3f}")
print(f"Max F2-Score achieved at this optimal threshold: {max_f2_score:.4f}")

# --- DIAGNOSTIC: Analyze F2-score distribution ---
print("\n--- Diagnostic: F2-Score values across thresholds ---")
print(pd.Series(f2_scores_at_thresholds).describe())
print(f"Number of thresholds with F2 > 0.01: {np.sum(np.array(f2_scores_at_thresholds) > 0.01)}")
print("---------------------------------------")

# Re-evaluate selected model with optimal threshold on test set for full report
y_pred_optimal_threshold_test = (y_pred_proba_final_model_test >= optimal_threshold).astype(int)

# --- Final Evaluation Output (Matching Desired Format) ---
print("\n--- Final Evaluation Summary ---")

# 1. Train Metrics (using default 0.5 threshold on resampled training data)
print("Train Metrics (threshold=0.5):")
y_train_pred_proba_default = final_best_model_for_prediction.predict_proba(X_train_resampled_scaled)[:, 1]
y_train_pred_default = (y_train_pred_proba_default >= 0.5).astype(int)
print(classification_report(y_train_resampled, y_train_pred_default, zero_division=0))

# 2. Validation/Test F2 Score (using optimal threshold on test data)
print(f"Validation F2 Score: {fbeta_score(y_test, y_pred_optimal_threshold_test, beta=2, zero_division=0):.10f}")

# Plot F2-score vs. threshold (for understanding, not part of strict final output format)
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f2_scores_at_thresholds, marker='o', linestyle='-', markersize=4, label='F2-Score')
plt.title(f'F2-Score vs. Classification Threshold for {best_model_name_for_threshold}')
plt.xlabel('Threshold')
plt.ylabel('F2-Score')
plt.axvline(x=optimal_threshold, color='r', linestyle='--', label=f'Optimal Threshold: {optimal_threshold:.3f} (F2={max_f2_score:.4f})')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot Precision-Recall Curve (for more context)
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba_final_model_test)
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve for {best_model_name_for_threshold}')
plt.grid(True)
plt.legend()
plt.show()


# Display Confusion Matrix for optimal threshold (for understanding)
print(f"\nConfusion Matrix for {best_model_name_for_threshold} at Optimal Threshold {optimal_threshold:.3f}:")
cm = confusion_matrix(y_test, y_pred_optimal_threshold_test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted No Default', 'Predicted Default'],
            yticklabels=['Actual No Default', 'Actual Default'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()


# Store the final best model and optimal threshold for actual prediction on the unlabeled validation set
# (These will be used in the final prediction step, which is next)
# global final_best_model_for_prediction, optimal_threshold
# final_best_model_for_prediction = final_best_model_for_prediction
# optimal_threshold = optimal_threshold
```
